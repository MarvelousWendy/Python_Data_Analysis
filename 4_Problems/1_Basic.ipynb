```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# # Job Postings Count
len(df)

# # Job information 
df.info()
```

```python
# Calculate Total Compensation
# 1- Annualized hourly salary
df["salary_from_hour"] = df["salary_hour_avg"] * 2080
# 2- Combine yearly + converted hourly
df["salary_annual"] = df["salary_year_avg"].fillna(df["salary_from_hour"])
# 3- 
comp_by_title = (
    df.groupby("job_title_short")["salary_annual"]
    .mean()
    .sort_values(ascending=False)
)
print(comp_by_title.head())
```

```python
# Calculate Monthly Salary
# Convert hourly salary to annual equivalent
df["salary_from_hour"] = df["salary_hour_avg"] * 2080

# Use yearly salary if available, otherwise converted hourly
df["salary_annual"] = df["salary_year_avg"].fillna(df["salary_from_hour"])

df["salary_monthly"] = df["salary_annual"] / 12

# df[["salary_year_avg", "salary_hour_avg", "salary_monthly"]].head()

average_monthly_salary = df["salary_monthly"].mean(skipna=True)
print(f"Average Monthly Salary: ${average_monthly_salary:,.2f}")
```
