```python
# Job Title Contains
job_title = 'Data Scientist'
keyword = 'Data'


def is_contains_keyword(keyword, title):
    return keyword in title


print(is_contains_keyword(keyword, job_title))
```

```python
# Average Salary
salaries = [95000, 120000, 105000, 90000, 130000]


def average_salary(salary_list):
    return sum(salary_list) / len(salary_list)


print(average_salary(salaries))
```

```python
# Salary Statistics
salaries = [95000, 120000, 105000, 90000, 130000]


def salary_stats(salary_list):
    return {
        'average': sum(salary_list) / len(salary_list),
        'min': min(salary_list),
        'max': max(salary_list)
    }


print(salary_stats(salaries))
```

```python
# Job Posting Summary
job_postings = [
    {'title': 'Data Scientist', 'location': 'New York', 'salary': 95000},
    {'title': 'Data Analyst', 'location': 'San Francisco', 'salary': 85000},
    {'title': 'Machine Learning Engineer', 'location': 'New York', 'salary': 115000}
]


def summarize_job_postings(postings):
    summary = {}
    for posting in postings:
        location = posting['location']
        if location not in summary:
            summary[location] = {'count': 0, 'average_salary': 0}
        summary[location]['count'] += 1
        summary[location]['average_salary'] += posting['salary']
    for location in summary:
        summary[location]['average_salary'] /= summary[location]['count']
    return summary


print(summarize_job_postings(job_postings))
```
