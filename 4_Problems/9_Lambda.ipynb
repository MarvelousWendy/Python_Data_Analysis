```python
# Calculate Average Salary
salaries = [95000, 120000, 105000, 90000, 130000]
def average_salary(x): return sum(x) / len(x)


print(average_salary(salaries))
```

```python
# Filter Job Titles
job_titles = ['Data Scientist', 'Data Engineer',
              'Machine Learning Engineer', 'Data Analyst']
filtered_titles = list(filter(lambda title: 'Data' in title, job_titles))
print(filtered_titles)
```

```python
# Filter Remote Python Jobs
job_postings = [
    {'title': 'Data Scientist', 'skills': ['Python', 'SQL'], 'remote': True},
    {'title': 'Data Analyst', 'skills': ['Excel', 'SQL'], 'remote': False},
    {'title': 'Machine Learning Engineer', 'skills': [
        'Python', 'TensorFlow'], 'remote': True},
    {'title': 'Software Developer', 'skills': ['Java', 'C++'], 'remote': True}
]
remote_python_jobs = list(
    filter(lambda job: 'Python' in job['skills'] and job['remote'], job_postings))
print(remote_python_jobs)
```
