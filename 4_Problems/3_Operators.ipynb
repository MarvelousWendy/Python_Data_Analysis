```python
# Compare Salaries
avg_salary_analyst = 70000
avg_salary_scientist = 85000
print(avg_salary_scientist > avg_salary_analyst)
```

```python
# Calculate Average Salary
min_salary = 50000
max_salary = 120000
average_salary = (min_salary + max_salary) / 2
print(average_salary)
```

```python
# Determine Python Experience
python_experience_years = 3
has_python_experience = python_experience_years >= 2
print(has_python_experience)
```

```python
# Determine Job Skill Level
tools_known = 3
if tools_known >= 5:
    skill_level = 'Expert'
elif tools_known >= 3:
    skill_level = 'Intermediate'
else:
    skill_level = 'Beginner'
print(skill_level)
```

```python
# Determine Job Title
job_experience = 4
sql_experience = 2
if job_experience >= 5 and sql_experience >= 3:
    job_title = 'Senior Data Scientist'
elif job_experience >= 3 and sql_experience >= 1:
    job_title = 'Data Scientist'
else:
    job_title = 'Junior Data Scientist'
print(job_title)
```
