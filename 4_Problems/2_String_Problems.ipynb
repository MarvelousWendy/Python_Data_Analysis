```python
# Cover to Uppercase
job_title = 'Data Scientist'
job_title.upper()
print(job_title)

# Convert to lowercase
job_title = 'DATA ENGINEER'
job_title.lower()
print(job_title)

# Replace Substring
job_title = 'Database Manager'
job_title = job_title.replace('Manager', 'Administrator')
print(job_title)

# Concatenate Strings
seniority_level = 'Senior'
job_title = 'Data Analyst'
full_title = seniority_level + ' ' + job_title
print(full_title)

# Check String Length
job_title = 'Data Scientist'
length = len(job_title)
print(length)
```

```python
# Concatenate Strings
position = 'Data Scientist'
company = 'DataWiz Inc.'
full_title = position + ' at ' + company
print(full_title)
```

```python
# Format with f-strings
position = 'Data Scientist'
company = 'DataWiz Inc.'
full_title = f'{position} at {company}' 
print(full_title)
```

```python
# Format with str.format()
position = 'Data Scientist'
company = 'DataWiz Inc.'
full_title = '{} at {}'.format(position, company)
print(full_title)
```

```python
# Format with %-formatting
position = 'Data Scientist'
company = 'DataWiz Inc.'
full_title = '%s at %s' % (position, company)
print(full_title)
```
