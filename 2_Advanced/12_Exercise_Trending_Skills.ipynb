Question

- How are skills trending for Data Analysts on a monthly basis.


load data


```python
# Importing Libraries
import ast  # Abstract Syntax Tree
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(
    lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

Trends for Data Analyst Jobs


1. For this we're going to only be looking at data analyst jobs. So we'll create a copy of our original DataFrame that filters only for Data Analyst roles.


```python
df_trends = df[df['job_title_short'] == 'Data Analyst'].copy()
```

2. Extract the month from the `job_posted_date`.


```python
df_trends['job_posted_month_no'] = df_trends['job_posted_date'].dt.month
```

3. Explode the `job_skills` column.


```python
df_trends_explode = df_trends.explode('job_skills')
```

4. Let's pivot the data to get the count (or size) of each skill for each month.


```python
df_trends_pivot = df_trends_explode.pivot_table(
    index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_trends_pivot
```

5. We need to sort the skills to be in order of count to make it easier to plot the top X.


```python
# Create a total row
df_trends_pivot.loc['Total'] = df_trends_pivot.sum()

# Order columns by total count
df_trends_pivot = df_trends_pivot[df_trends_pivot.loc['Total'].sort_values(
    ascending=False).index]

# Drop total row
df_trends_pivot = df_trends_pivot.drop('Total')

df_trends_pivot
```

(Optional) months as names, not numbers, so let's change this.


```python
df_trends_pivot = df_trends_pivot.reset_index()
df_trends_pivot['job_posted_month'] = df_trends_pivot['job_posted_month_no'].apply(
    lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_trends_pivot = df_trends_pivot.set_index('job_posted_month')
df_trends_pivot = df_trends_pivot.drop(columns='job_posted_month_no')

df_trends_pivot
```
